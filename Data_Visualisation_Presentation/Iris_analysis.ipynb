# Petal : Petals are modified leaves that surround the reproductive parts of flowers.

```python
# Bar Chart
plt.figure(figsize=(10, 6))
sns.barplot(x="Species", y="SepalLengthCm", data=df,palette=['orange','red','blue'])
plt.title("Average Sepal Length per Species")
plt.xlabel("Species")
plt.ylabel("Average Sepal Length (cm)")
plt.show()


```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x="Species", y="SepalWidthCm", data=df,palette=['green','yellow','pink'])
plt.title("Average Sepal Width per Species")
plt.xlabel("Species")
plt.ylabel("Average Width Length (cm)")
plt.show()
```

# Bar Chart: Shows the average Sepal Length for each species.

```python
# Line Chart
df_setosa = df[df["Species"] == "Iris-setosa"]
plt.figure(figsize=(10, 6))
plt.plot(df_setosa["Id"], df_setosa["SepalLengthCm"], label="Sepal Length")
plt.plot(df_setosa["Id"], df_setosa["SepalWidthCm"], label="Sepal Width")
plt.plot(df_setosa["Id"], df_setosa["PetalLengthCm"], label="Petal Length")
plt.plot(df_setosa["Id"], df_setosa["PetalWidthCm"], label="Petal Width")
plt.title("Sepal Length and Width for Iris-setosa")
plt.xlabel("Id")
plt.ylabel("Measurement (cm)")
plt.legend()
plt.show()
```

# Line Chart: This line chart shows the trend of Sepal Length and Sepal Width for Iris-setosa across different instances (Id). This helps to observe variations within this species.

```python
# Histogram
plt.figure(figsize=(10, 6))
sns.histplot(df["PetalLengthCm"], kde=True)
plt.title("Distribution of Petal Length")
plt.xlabel("Petal Length (cm)")
plt.ylabel("Frequency")
plt.show()
```

# Histogram: The histogram plots the distribution of Petal Length across all instances and species in the dataset. The KDE (Kernel Density Estimate) overlay provides a smoothed curve to see the distribution pattern more clearly.


# Next we have some Box Plots


```python
Image(url="https://www.onlinemathlearning.com/image-files/box-plot.png", width=300, height=200) 
```

```python
# Box Plot
plt.figure(figsize=(16, 8))

# Sepal Length
plt.subplot(2, 2, 1)
sns.boxplot(x="Species", y="SepalLengthCm", data=df,palette=["green"])
plt.title("Box plot of Sepal Length by Species")

# Sepal Width
plt.subplot(2, 2, 2)
sns.boxplot(x="Species", y="SepalWidthCm", data=df, palette=["orange"])
plt.title("Box plot of Sepal Width by Species")

# Petal Length
plt.subplot(2, 2, 3)
sns.boxplot(x="Species", y="PetalLengthCm", data=df,palette=["yellow"])
plt.title("Box plot of Petal Length by Species")

# Petal Width
plt.subplot(2, 2, 4)
sns.boxplot(x="Species", y="PetalWidthCm", data=df,palette=["blue"])
plt.title("Box plot of Petal Width by Species")

plt.tight_layout()
plt.show()
```

# Box Plot of Sepal Length by Species: Shows the distribution of Sepal Length for each species.
# Box Plot of Sepal Width by Species: Shows the distribution of Sepal Width for each species.
# Box Plot of Petal Length by Species: Shows the distribution of Petal Length for each species.
# Box Plot of Petal Width by Species: Shows the distribution of Petal Width for each species.
